```python
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
from selenium import webdriver

driver = webdriver.Chrome(executable_path = 'Users/User/Downloads/chromedriver')

url = 'https://www.facebook.com'

driver.get(url)
```

```python
url = 'https://www.facebook.com/marketplace/toronto/vehicles?maxPrice=4000&exact=false'

driver.get(url)
```

```python
elements = driver.find_elements(By.CLASS_NAME,"x3ct3a4")

df = pd.DataFrame()

#df.columns = ['line_1']

for ele in elements:
    new_row = pd.Series(ele.text)
   
    df = df.append(new_row.str.split('\n', expand=True))
    
df = df.set_axis(['Price', 'Title', 'Location','Mileage','Extends'], axis=1)
df
```
